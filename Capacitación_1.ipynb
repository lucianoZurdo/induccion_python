### Variables y variables

En python no hace falta decir que tipo de variable es, solo tenemos que igualar al valor y python mismo se dará cuenta de su tipo, por lo general tenemos 4 tipos de variables muy muy generales que se suelen usar tooodo el tiempo: numerica, string (o cadena de texto), lista y diccionario.

Como adicional la función para imprimir cosas en pantalla se llama "print()"

```python
# Declaración de variables
a = 14
b = "hola soy una cadena"
x = True
c = [1,2,3,"hola soy un objeto dentro de la cadena"]
d = {'hola soy la llave' : 'hola soy el valor', 'hola soy otra llave' : 'soy otro valor'}

# imprimimos en pantalla
print('La variable a es: ',a,'. \nMientras que la variable b:"',b,'".\nLa variable C: ',c,'. \nPor ultimo la variable d: ',d);
```

Es importante saber que cada tipo de variable tiene funciones asociadas a si misma por su clase, es decir si es numérica, string, lista o diccionario tendrá asociadas distintas funciones a su tipo, por ejemplo:

```python
a = a**2                   # Elevamos al cuadrado, no es una función estrictamente como tal pero solo podemos hacerlo con numeros.
b = b.upper()              # Volvemos mayuscula al texto
c.append('nuevo item')     # Agregamos un nuevo item
d = d.keys()               # Obtenemos las llaves del diccionario

# Ahora printeamos el resultado
print('La variable a es: ',a,'. \nMientras que la variable b:"',b,'".\nLa variable C: ',c,'. \nPor ultimo la variable d: ',d);
```

### Bucles


En python sucede algo especial con todas aquellas variables que pueden ser partidas en pedazitos y esto es que son lo que denominamos "iterables" es decir que podemos ir separando en partecitas a estas variables, pooor ejemplo:

```python
palabra = 'esto es una palabra'

print('Iteramos por las letras de la frase')

lista = ['abc','cde','efg'];

for letra in lista:
  print('Letra:',letra)
```

Cositas importantes de la sintaxis, cuando hacemos un bucle tenemos que SI O SI identar el texto, (identar es llevar adelante el texto y no dejarlo pegadito a la izquierda). El texto se identa con cuatro espacios o un tab.

Ahora presten atencio en que se pudo partir en pedazitos a esta variable, esto es CLAVE para un monton de cosas en python, porque POR LO GENERAL las cosas vienen en variables que son iterables.

Las listas y los diccionarios son otro tipo de objetos iterables

```python
diccionario = {'a':1, 'b':2, 'c':3}
lista       = ['d','e','f']

print('cosas del diccionario')
for claves in diccionario:
  print(claves)

print('cosas de la lista')
for items in lista:
  print(items)
```

Por lo general el 95% de los casos se resuelven con un bucle de tipo "for" como el que vimos arriba, pero también hay otro tipo de bucle llamado "while", este tipo de bucle se caracteriza por funcionar con una condición lógica.

Esto lo vemos en la siguiente sección donde metemos también condiciones lógicas

### Condiciones lógicas

Una condición lógica es escencialmente una comparación entre variables, por ejemplo:

si a = 2 y b = 3 entonces: a < b (a es menor a b)

por lo cual eso nos daria un "True" de resultado, que quiere decir que esa condición que escribimos es verdad. Sin embargo podemos hacer lo contrario y poner:

a > b

Lo cual es Falso, por lo que python nos devolverá "False"

```python
a = 2
b = 3

# Si decimos que a es menor que b
print('Si decimos que a es menor que b:',a < b)

# Si decimos que a es mayor que b
print('Si decimos que a es mayor que b:',a > b)
```

Entonces ahora conociendo que es una condición lógica podemos utilizar el while

```python
# Instaciamos la variable numerica a
a = 1

# Empezamos a hacer un bucle siempre que a sea menor a 5
while a < 5:
  print(a)
  a = a + 1 # incrementamos a en uno por cada vuelta que damos, es decir primero valdrá 1, luego 2, luego 3 y así
```

Como vemos el bucle se cumplio hasta que la condición lógica fue falsa, a partir de este punto el bucle termina.

### Funciones

Las funciones son unas maquinitas que podemos armar dentro de la programación para que hagan cosas por nosotros. 

Por lo general funcionan como un proceso, le damos inputs o entradas (esto es opcional) y obtenemos resultados (esto también es opcional) por ejemplo:

```python
def concatenador_palabras(input_1, input_2): # Fabricamos la función con sus inputs
  resultado = input_1 + ' ' + input_2        # Hacemos la cuenta que queremos o el proceso que deseamos
  return resultado                           # Devolvemos el valor calculado

print(concatenador_palabras('hola','amigos'))
```

Otro ejemplo con parámetros por defecto:

```python
def elevador_cuadrado(numero_1 = 1, numero_2 = 0):
  resultado = numero_1**numero_2;
  return resultado;

# Sin parametros
print(elevador_cuadrado())

# Con un parámetro
print(elevador_cuadrado(2))

# Con todos los parámetros
print(elevador_cuadrado(2,3))
```

### Clases

Las clases son objetos de el paradigma orientado a objetos que bien podemos hablar mucho sobre ellas, la posta es que no tenemos mucho tiempo asi que me voy a limitar a contarles como se escriben en este lenguaje

```python
class humano:

  # Metodo constructor del objeto
  def __init__(self, nombre, anios):
    self.nombre = nombre;  # atributos del objeto humano
    self.anios  = anios;   # atributos del objeto humano

  # Funciones que puede hacer un humano
  def trabajar(self, tarea):
    print('El humano', self.nombre, 'esta haciendo', tarea);

  def hablar_con_amigos(self, amigo):
    print('El humano', self.nombre, 'habla con',amigo)
```

```python
# Instanciamos un objeto de la clase humano
fulanito = humano('cosme fulanito',38);

# Le decimos que haga una de sus funciones y le damos el parámetro
fulanito.hablar_con_amigos('lucho')
```
